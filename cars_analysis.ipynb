```python
import pandas as pd
```

```python
car = pd.read_csv("file.csv")
```

```python
#To view DF, by default head shows 5 rows
car.head(2)
```

```python
#To check no of rows and columns , 432 rows and 15 colns
car.shape
```

```python
#Find all Null values present and then fill it with mean of that column
# isnull() output is true/false
car.isnull()
car.isnull().sum() #Total no of nulls in every column
```

```python
#To replace Cylinder column nulls with mean, inplace = True will make changes to actual dataframe
car['Cylinders'].fillna(car['Cylinders'].mean(), inplace = True)
```

```python
#Value counts,like group by in SQL
#Check different kinds of make(coln name) and their count in dataset
car['Make'].value_counts()
```

```python
#Filtering, Show all records where origin is asia or europe,isin() pass values as a list
car[car['Origin'].isin(['Asia','Europe'])]
```

```python
##Remove records where weight > 4000
#car['Weight'] > 4000 # output in boolean format
car[car['Weight'] > 4000] #Filtered output of whole DF
car[~(car['Weight'] > 4000)] #~ sign to remove that
```

```python
##change value of a column by 3
car['MPG_City']= car['MPG_City'].apply(lambda x:x+3)
```

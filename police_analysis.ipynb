```python
import pandas as pd
```

```python
#skipinitialspace used as map function returned NAN
data= pd.read_csv('police.csv',skipinitialspace=True)
```

```python
data
```

```python
#Remove columns containing missing values
#data.isnull().sum() # This will give total null values in each column
# Drop function to remove rows or columns, if parameter index is used row will be removed
data.drop('country_name',axis='columns', inplace = True)
```

```python
# Filtering + Value Counts, For speeding were men or women stopped more
data[data.violation == 'Speeding'] # to display whole DF which satisfy this condition
# data['stop_date'] to display only stop_date coln
data[data.violation=='Speeding'].driver_gender.value_counts()
```

```python
#does gender affect who gets searched during a stop
data[data.search_conducted==True].driver_gender.value_counts()
# Other way is using groupby and sum
data.groupby('driver_gender').search_conducted.sum()
```

```python
#mean stop_duration
# stop_duration column is string, first convert to integer
# to see unique values  of stop_duration data['stop_duration'].value_counts()
#data['stop_duration'].value_counts()
#data['stop_duration']= data['stop_duration'].map({'0-15 Min' : 7.5,'16-30 Min' : 24, '30+ Min' : 45})
#above step converts to mean
#data.isnull().sum()
data['stop_duration'].mean()
```

```python
#Find all null values in data
data.isnull() # whole dataframe as output , with True/False indicating if that particular cell contains null or not
data.isnull().sum() # No of nulls for each columns
data.notnull().sum() # No of not nulls for each columns
```

```python
#Rename column weather to weathe_condition
data.rename(columns ={'Weather':'Weather_Condition'},inplace=True)
```

```python
data.head()
```

```python
#mean visibility
data.Visibility_km.mean()
data['Visibility_km'].mean() #if there is space between column name square brackets have to be used, else above syntax will work fine too
```

```python
#standard devaition of pressure column
data['Press_kPa'].std()
```

```python
#variance of relative humidity
data['Rel Hum_%'].var()
```

```python
#all instances when snow was recorded
data['Weather_Condition'].value_counts()
data[data['Weather_Condition']=='Snow']
#Above 2 will fetch exact match, better approach will be str.contains()
data[data['Weather_Condition'].str.contains('Snow')].head(25)
```

```python
# Wind speed above 24 and visibility is 25
data[(data['Wind Speed_km/h']>24) & (data['Visibility_km']==25)]
```

```python
#mean,max,min value of each column against weather_condition column
data.groupby('Weather_Condition').mean()
data.groupby('Weather_Condition').max()
data.groupby('Weather_Condition').min()
```

```python
#Weather condiition is Fog
data[data['Weather_Condition']=='Fog'] #150 rows
data[data['Weather_Condition'].str.contains('Fog')].tail(10) #420 rows
```

```python
#weather is clear or visibility above 40
data[(data['Weather_Condition']=='Clear') | (data['Visibility_km']>40)]
```

```python
#weather is clear and relative humidity more than 50 or visibility > 40
data[((data['Weather_Condition']=='Clear') & (data['Rel Hum_%']>50)) | (data['Visibility_km']>40)]
```
